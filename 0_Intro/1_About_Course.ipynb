**⚠️ Note:** I will be reviewing concepts if you haven't touched SQL in a while but, we'll be moving quickly. This course isn't suitable for those without the basics down (see below for required concepts to know)

### 💽 What SQL Knowledge do you need?

- 🔍 **Data Retrieval**: `SELECT`, `FROM`, `WHERE`, `ORDER BY`, `LIMIT`
- 📊 **Aggregation & Grouping**: `COUNT()`, `SUM()`, `AVG()`, `GROUP BY`, `HAVING`
- 🔗 **Joins & Unions**: `INNER/FULL JOIN`, `LEFT/RIGHT JOIN`, `UNION`
- 🧠 **Logic, Dates & Conditions**: `AND`, `OR`, `NOT`, `IN`, `LIKE`, `BETWEEN`, `CASE`, `::DATE`
- 🌀 **Subqueries & CTEs**: `(SELECT ...)`, `WITH cte_name AS (...)`
- 🛠 **Data & Schema Control**: `INSERT`, `UPDATE`, `DELETE`, `CREATE`, `ALTER`, `DROP`

### 🧮 What Math knowledge do you need to know?

- ➕ **Arithmetic**: `+`, , , `/`
- 🔄 **Order of Ops**: Follows `PEMDAS`
- 📈 **Algebra**: `price * quantity`
- 📊 **Basic Stats**: `AVG()`, `COUNT()`, `SUM()`
- 📉 **Percentages**: `(value / total) * 100`

🎫 **If you have a high school education, you likely have all the math knowledge needed.**

## Course Outline

### 1️⃣ First Half

- Intro: We'll get you set up and comfortable with the database we'll be using throughout the course
- Pivoting Data with Case Statements: Building on that foundation, we'll transform and analyze data using both basic and advanced statistical methods
- Intermediate Date & Time functions: Next up, we'll tackle those tricky date-time values, showing you the best ways to clean and filter them
- Window Functions: With our solid foundation, we'll dive into everyone's favorite topic! Starting with the basics and working up to complex aggregations

### 2️⃣ Second Half

- Local setup of PostgreSQL w/ PGAdmin: Now that we've mastered the concepts, let's set up your own environment with Postgres and DBeaver
- Views: With your local environment ready, we'll build your first View and incorporate it into a portfolio project
- Data Cleaning: After this, you'll learn all my popular functions to transform messy data into clean datasets
- Query Optimization: To wrap up our technical journey, we'll explore how to use powerful tools like `EXPLAIN` to make your queries super fast

### 3️⃣ Final Project

- Finally, we'll put everything together in a complete project that showcases your new skills
- You'll tackle 3 real business questions that data analysts face in their day-to-day work
- We'll share it all on GitHub to help build your portfolio
